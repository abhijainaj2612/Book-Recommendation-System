## Data Preprocession

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df['Title'].duplicated().sum()
```

```python
df = df.drop_duplicates(subset='Title', keep='first')
```

```python
df['Title'].duplicated().sum()
```

```python
df.duplicated().sum()
```

```python
df.shape
```

```python
df.info()
```

```python
df['Author']=df['Author'].apply(lambda x: list(set(word.strip() for word in x.split(','))))
```

```python
df['Genres']=df['Genres'].apply(lambda x: list(word.strip() for word in x.split(',')))
```

```python
df.info()
```

```python
df.head()
```

```python
df["Description"] = df["Description"].apply(lambda x: x.split())
```

```python
df["Description"] = df["Description"].apply(lambda x : [i.replace(" ","") for i in x])
df["Author"] = df["Author"].apply(lambda x : [i.replace(" ","") for i in x])
df["Genres"] = df["Genres"].apply(lambda x : [i.replace(" ","") for i in x])
```

```python
df["tags"] = df["Description"] + df["Author"] + df["Genres"]
```

```python
new_df = df[["Title","Author", "tags"]]
```

```python
new_df["tags"] = new_df["tags"].apply(lambda x: " ".join(x))
```

```python
new_df["tags"] = new_df["tags"].apply(lambda x: x.lower())
```

```python
new_df.iloc[0]["tags"]
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y =[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df["tags"] = new_df["tags"].apply(stem)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
    max_features=5000,
    stop_words='english'
)

tfidf_matrix = tfidf.fit_transform(new_df['tags']).toarray()
```

```python
tfidf_matrix
```

```python
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(tfidf_matrix)
```

```python
similarity
```

```python
def recommend(movie):
    if movie not in new_df['Title'].values:
        return "Book not found. Please try another title."
    index = new_df[new_df['Title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    recommendations = []
    for i in distances[1:6]:
        recommendations.append({
            "title": new_df.iloc[i[0]].Title,
            "author": new_df.iloc[i[0]].Author,
            "similarity_score": i[1]
        })
    return recommendations
```

```python
recommend('Gandhi')
```
