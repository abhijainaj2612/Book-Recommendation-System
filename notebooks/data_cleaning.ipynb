## Data Preprocession

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df['Title'].duplicated().sum()
```

```python
df = df.drop_duplicates(subset='Title', keep='first')
```

```python
df['Title'].duplicated().sum()
```

```python
df.duplicated().sum()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Author']=df['Author'].apply(lambda x: ",".join(sorted(set(word.strip() for word in x.split(',')))))
```

```python
df['Num_Ratings'] = df['Num_Ratings'].astype(str).str.replace(' ratings', '', regex=False).str.replace(',', '', regex=False)
df['Num_Ratings'] = pd.to_numeric(df['Num_Ratings'], errors='coerce')
```

```python
df['Num_Reviews'] = df['Num_Reviews'].astype(str).str.replace(' reviews', '', regex=False).str.replace(',', '', regex=False)
df['Num_Reviews'] = pd.to_numeric(df['Num_Reviews'], errors='coerce')
```
